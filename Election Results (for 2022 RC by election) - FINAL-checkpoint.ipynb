### Patcham ward results - including a by election in 2021

```python
years = [2011, 2015, 2019, 2021]
con = [41, 47, 45, 42]
grn = [26, 15, 24, 36]
lab = [19, 29, 25, 18]
lib = [4, 6, 0, 4]
kip = [0, 12, 0, 1]
```

```python
df2 = pd.DataFrame({'Year':years, 'Con':con, 'Green':grn, 'Labour':lab, 'LD':lib, 'UKIP':kip})
```

```python
df2
```

```python
sns.lineplot(x='Year', y='Con', data=df2, color='blue', linestyle="dashed").set(ylabel='Vote (%)')
sns.lineplot(x='Year', y='Green', data=df2, color='green', linestyle="dashed")
sns.lineplot(x='Year', y='Labour', data=df2, color='red', linestyle="dashed")
sns.lineplot(x='Year', y='LD', data=df2, color='orange', linestyle="dashed")
sns.lineplot(x='Year', y='UKIP', data=df2, color='purple', linestyle="dashed")
```

### Rottingdean Coastal ward results, 2011-19

```python
years = [2011, 2015, 2019]
con = [53, 54, 35]
grn = [22, 13, 16]
lab = [17, 24, 27]
lib = [8, 9, 13]
kip = [0, 10, 0]
ind = [0, 0, 41]
```

```python
df3 = pd.DataFrame({'Year':years, 'Con':con, 'Green':grn, 'Labour':lab, 'LD':lib, 'UKIP':kip, 'Ind':ind})
```

```python
sns.lineplot(x='Year', y='Con', data=df3, color='blue').set(ylabel='Vote (%)')
sns.lineplot(x='Year', y='Green', data=df3, color='green')
sns.lineplot(x='Year', y='Labour', data=df3, color='red')
sns.lineplot(x='Year', y='LD', data=df3, color='orange')
sns.lineplot(x='Year', y='UKIP', data=df3, color='purple')
sns.lineplot(x='Year', y='Ind', data=df3, color='gray')
```
